```python
#csv -> comma separated values file
#other formats : .read_parquet,read_excel
coffee = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/complete-pandas-tutorial/refs/heads/master/warmup-data/coffee.csv')
```

```python
coffee.head() #->access first 5 rowa
#coffee.tail() -> access last 5 rows
#coffe.sample() -> scattered values
```

```python
coffee
```

```python
coffee.sample(10,random_state=1)
#using random_state = 1 gives us only one shuffled version
```

```python
coffee.loc[[0,1,2],["Day","Units Sold"]]
#coffee.loc[[1,2,4]]
#coffee.loc[1]
#coffee.loc[2:3]
#upper index inclusive
```

```python
coffee.iloc[[1,2,4],[0,2]] #instead of column name,
#we are using column index
#upper index not inclusive
```

```python
#changing value
coffee.loc[0:5,"Units Sold"] = 10
coffee.loc[[0,1,2]]
```

```python
coffee["Units Sold"]
```

```python
#sorting
coffee.sort_values(["Units Sold","Coffee Type"],ascending = [0,1])
#coffee.sort_values("Units Sold",ascending = False)
```
